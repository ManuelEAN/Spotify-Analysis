```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importing dataset and cleaning it

```python
df = pd.read_csv(r'C:\Users\admin\Documents\DataAnalystProjects\Spotify\tracks.csv')
#Drop rows with NA values
df.dropna(inplace=True)
df.head()
```

```python
df.head()
```

```python
#Top 10 songs with worst popularity
sort_df = df.sort_values('popularity', ascending = True).head(10)
sort_df.head()
```

```python
#Statistics about dataset
df.describe().transpose()
```

# Top popular songs

```python
most_popular = df[df['popularity'] > 90]
most_popular.sort_values('popularity', ascending = False).head(10)
```

# Changing the index column in the dataset

```python
#Setting another dataframe to work on it with different index

df_sec = df
df_sec['release_date'] = df_sec['release_date'].astype('datetime64')  #changing column to datetime
df_sec.set_index('release_date', inplace = True)  #Setting release_date column as index column
df_sec.head()
```

# Applying function to columns in dataset

```python
#Convert the duration from ms to seconds
df_sec['duration'] = df_sec['duration_ms'].apply(lambda x: round(x/1000))
df_sec.drop('duration_ms', inplace = True , axis=1)
df_sec.head()
```

# Heatmap of Correlations between variables

```python
df_heat = df.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(df_heat, annot =True , fmt='.1g', vmin=-1, center=0 , cmap = 'inferno' , linewidths=1 , linecolor = 'Black')
heatmap.set_title('Correlation between variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
plt.show()
```

# Correlation between Loudness and Popularity Chart

```python
plt.figure(figsize=(10,6))
sns.regplot(data = df , y = 'loudness', x = 'popularity' , color='c')
plt.title('Loudness vs Popularity Correlation')
plt.show()
```

# Creating a new column from filtering time column

```python
#Creating a new column of the Year of the release
df['dates'] = df.index.get_level_values('release_date')
df['dates'] = pd.to_datetime(df.dates)
df['years'] = df.dates.dt.year
df.head(3)
```

# Songs per year Chart

```python
#Graph of songs released by years

sns.displot( df.years , discrete = True, aspect = 2, height = 5, kind='hist', color ='b')
plt.title('Number of songs per year')
plt.show()
```

# Duration of the songs vs Year Bars Chart

```python
total_dur = df.duration
fig, ax = plt.subplots(figsize=(18,7))
sns.barplot(x = df.years , y = df.duration, ax = ax, errorbar = None)
plt.title('Year vs Duration ')
plt.xticks(rotation=90)
plt.show()
```

# Duration of the songs vs Year Line Chart

```python
total_dur = df.duration
sns.set_style(style='whitegrid')
fig_dims = (10,5)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.lineplot(x = df.years , y = df.duration, ax = ax , color='c').set(title='Year vs Duration')
plt.xticks(rotation=60)
plt.show()
```
